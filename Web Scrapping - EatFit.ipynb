```python
from autoscraper import AutoScraper
```

```python
eatfit_url = 'http://eatfit.in/curefoodsordernow'
wanted_list = ['₹ 83','Sabudana Khichdi','Nutritional Info.']
```

```python
scraper=AutoScraper()
result=scraper.build(eatfit_url,wanted_list)
```

```python
print(result)
```

```python
scraper.get_result_similar(eatfit_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_btjg':'Price','rule_xr6b':'Title'})
scraper.keep_rules(['rule_btjg','rule_xr6b'])
```

```python
results=scraper.get_result_similar(eatfit_url,group_by_alias=True)
```
